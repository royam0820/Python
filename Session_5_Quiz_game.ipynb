# Jeu de Quiz sur Harry Potter

```python
# Quiz Harry Potter en Python pour Google Colab
# Ce quiz teste les connaissances sur la saga Harry Potter avec des questions √† choix multiples
# Inclut des commentaires de type IA pour les r√©ponses des utilisateurs

def display_welcome_message():
    print("Bienvenue au Quiz Harry Potter !")
    print("Testez vos connaissances sur le monde des sorciers.")
    print("Choisissez la bonne r√©ponse (1 √† 4) pour chaque question.\n")

def get_questions():
    # Liste de questions avec options, r√©ponse correcte et commentaires
    return [
        {
            "question": "Comment s'appelle le hibou d'Harry Potter ?",
            "options": [
                "1. Hedwig",
                "2. Errol",
                "3. Pigwidgeon",
                "4. Crookshanks"
            ],
            "correct": 1,
            "correct_feedback": "G√©nial ! Hedwige est en effet la chouette fid√®le de Harry.",
            "incorrect_feedback": "Oh non, c'est faux ! La chouette d'Harry est Hedwige."
        },
        {
            "question": "Quelle maison de Poudlard est connue pour valoriser le courage ?",
            "options": [
                "1. Hufflepuff",
                "2. Ravenclaw",
                "3. Gryffindor",
                "4. Slytherin"
            ],
            "correct": 3,
            "correct_feedback": "Bravo ! √Ä Gryffondor, tout est une question de courage.",
            "incorrect_feedback": "Pas tout √† fait ! Gryffindor est la maison du courage."
        },
        {
            "question": "De quoi est fait le noyau de la baguette d'Harry ?",
            "options": [
                "1. Dragon heartstring",
                "2. Unicorn hair",
                "3. Phoenix feather",
                "4. Veela hair"
            ],
            "correct": 3,
            "correct_feedback": "Fantastique ! La baguette de Harry a un c≈ìur en plume de ph√©nix.",
            "incorrect_feedback": "Oups ! La bonne r√©ponse est 3 - plume de ph√©nix."
        },
        {
            "question": "Qui est le directeur de Poudlard lorsque Harry arrive pour la premi√®re fois ?",
            "options": [
                "1. Severus Snape",
                "2. Albus Dumbledore",
                "3. Minerva McGonagall",
                "4. Dolores Umbridge"
            ],
            "correct": 2,
            "correct_feedback": "Excellent ! Albus Dumbledore est le directeur.",
            "incorrect_feedback": "C'est faux ! Albus Dumbledore √©tait directeur √† l'√©poque."
        },
        {
            "question": "Quel sort est utilis√© pour invoquer des objets ?",
            "options": [
                "1. Expelliarmus",
                "2. Lumos",
                "3. Accio",
                "4. Alohomora"
            ],
            "correct": 3,
            "correct_feedback": "Parfait ! Accio est le charme d'invocation.",
            "incorrect_feedback": "Incorrect¬†! Accio est utilis√© pour invoquer des objets."
        }
    ]

def get_user_answer(question_index, question):
    print(f"Question {question_index + 1}: {question['question']}")
    for option in question['options']:
        print(option)  # Print each option on a new line
    while True:
        try:
            answer = int(input("Your answer (1-4): "))
            if 1 <= answer <= 4:
                return answer
            else:
                print("Veuillez saisir un nombre compris entre 1 et 4.")
        except ValueError:
            print("Entr√©e invalide ! Veuillez saisir un nombre compris entre 1 et 4.")


def provide_ai_feedback(is_correct, question):
    # Retour de type IA bas√© sur la justesse de la r√©ponse.
    if is_correct:
        print(question["correct_feedback"])
    else:
        print(question["incorrect_feedback"])
    print()  # Ligne vide pour plus de lisibilit√©

def display_final_score(score, total_questions):
    # Afficher le score final et un message amusant
    percentage = (score / total_questions) * 100
    print(f"\nQuiz termin√©¬†! Votre score¬†: {score}/{total_questions} ({percentage:.1f}%)")
    if percentage == 100:
        print("Tu es un vrai champion de Poudlard ! üèÜ")
    elif percentage >= 70:
        print("Bravo, tu es un sorcier talentueux ! ü™Ñ")
    elif percentage >= 50:
        print("Pas mal, mais vous aurez peut-√™tre besoin d'un petit rappel √† Poudlard ! üìö")
    else:
        print("On dirait que tu as besoin de revisiter le monde des sorciers ! ‚ö°")

def run_quiz():
    display_welcome_message()
    questions = get_questions()
    score = 0

    # Iterate through each question
    for i, question in enumerate(questions):
        user_answer = get_user_answer(i, question)
        is_correct = user_answer == question["correct"]
        if is_correct:
            score += 1
        provide_ai_feedback(is_correct, question)

    display_final_score(score, len(questions))

# Run the quiz
if __name__ == "__main__":
    run_quiz()
```
