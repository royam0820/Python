## 3. D√©veloppement du projet

### √âtapes pour construire le jeu :
1. **D√©finir la plage de nombres** : L'IA devinera un nombre entre 1 et 100.
2. **Faire des suppositions intelligentes** : L'IA utilisera une strat√©gie simple (comme la recherche binaire) pour deviner.
3. **Demander des indices** : L'utilisateur dira si la supposition est correcte, trop haute ou trop basse.
4. **Stocker les tentatives** : Utilisez une liste pour suivre les suppositions de l'IA.


### Code de base :

```python
# Jeu : L'IA devine votre nombre (Recherche Binaire)
import random

print("üéÆ Devine le Nombre avec l'IA !")
print("Pense √† un nombre entre 1 et 100. Je vais le deviner !")

low = 1
high = 100
essais = 0

while low <= high:
    devinette = (low + high) // 2 # // est utilis√© pour la division enti√®re.  Calcule le point m√©dian de la plage actuelle comme estimation de l'ordinateur.
    reponse = input(f"Est-ce {devinette} ? (Trop haut: 'H', Trop bas: 'B', Correct: 'O') ").upper()

    if reponse == 'O':
        print(f"Gagn√© en {essais} coups ! üéâ")
        break
    elif reponse == 'H':
        high = devinette - 1
        essais += 1
    elif reponse == 'B':
        low = devinette + 1
        essais += 1
    else:
        print("R√©ponse invalide. Utilise H, B ou O.")

if low > high:
    print("Tu as trich√© ? üòâ")
```

NB: Concepts de base :
- Variables : `bas`, `haut`, `tentatives`.
- Entr√©e utilisateur : `input()` pour interagir.

- Boucles : `while` pour continuer jusqu'√† ce que le joueur gagne.
- Module `random` : Introduit une biblioth√®que simple pour g√©n√©rer des nombres al√©atoires.

## POO - Les Concepts


**Les concepts**

- **Classe** : Un mod√®le ou un plan pour cr√©er des objets (ex. : un plan pour construire un jouet). Convention de nom de classe¬†: les noms de classe commencent g√©n√©ralement par une lettre majuscule.
- **Objet** : Une instance d‚Äôune classe, comme un jouet fabriqu√© √† partir du plan.
- **H√©ritage** : Une classe peut "h√©riter" des caract√©ristiques d‚Äôune autre classe, comme un chiot qui h√©rite des traits d‚Äôun chien.

### Voyant cela en code !

```python
# Un jeu de cr√©atures magiques avec combats (style Pok√©mon simple)

# D√©finition classe m√®re - ton blueprint ou mod√®le
class Creature:
    # initialisation
    def __init__(self, nom, force):
        self.nom = nom
        self.force = force

    # def m√©thode attaquer
    def attaquer(self):
        print(f"{self.nom} attaque avec une force de {self.force} !")

# Cr√©ation de l'objet
creature1 = Creature("Licorne", 10)

# Utilisation
creature1.attaquer()
```

```python
# D√©finition classe fille - la classe Dragon h√©rite de la classe Creature
class Dragon(Creature):
    # initialisation
    def __init__(self, nom, force, feu):
        super().__init__(nom, force)
        self.feu = feu

    # def m√©thode cracher_feu
    def cracher_feu(self):
        print(f"{self.nom} crache du feu avec une puissance de {self.feu} üî•")

# Cr√©ation des objets
creature1 = Creature("Licorne", 10)
dragon1 = Dragon("Drako", 15, 20)

# Utilisation
creature1.attaquer()
dragon1.attaquer()
dragon1.cracher_feu()
```
