# AI Prompt Helper

```python
import random
```

```python
def generate_prompt():
    # Listes d'éléments pour créer des prompts variés et amusants
    characters = ["un dragon magique", "un robot rigolo", "une licorne arc-en-ciel", "un pirate courageux", "un chat astronaute", "une sirène chanteuse"]
    actions = ["qui danse", "qui vole dans l'espace", "qui explore une jungle", "qui construit un château", "qui joue de la guitare", "qui nage dans un océan de bonbons"]
    settings = ["sur une planète en bonbons", "dans une forêt enchantée", "dans une ville futuriste", "sur un nuage géant", "dans un monde sous-marin", "au sommet d'une montagne arc-en-ciel"]
    styles = ["dans un style dessin animé", "comme une peinture à l'aquarelle", "en 3D réaliste", "comme un vitrail coloré", "dans un style pixel art", "comme une bande dessinée"]

    # Générer un prompt en combinant les éléments
    prompt = f"{random.choice(characters)} {random.choice(actions)} {random.choice(settings)} {random.choice(styles)}"
    return prompt
```

```python
def main():
    print("Bienvenue dans l'Aide aux Prompts d'Art IA !")
    print("Ce programme va t'aider à trouver des idées géniales pour créer des images avec l'IA.")

    while True:
        print("\nAppuie sur Entrée pour générer un nouveau prompt ou tape 'quitter' pour arrêter.")
        user_input = input()

        if user_input.lower() == "quitter":
            print("Merci d'avoir utilisé l'Aide aux Prompts d'Art IA ! À bientôt !")
            break

        # Générer et afficher un prompt
        prompt = generate_prompt()
        print("\nVoici ton prompt d'art IA :")
        print(prompt)
        print("\nEssaie ce prompt dans ton outil d'art IA préféré ou invente une histoire avec !")
```

```python
main()
```
