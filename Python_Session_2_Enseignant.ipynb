## 2. Prendre des d√©cisions avec If-Else
Parfois, nous voulons que notre programme fasse diff√©rentes choses selon la situation. C'est l√† que les instructions **if-else** entrent en jeu !

### Comment √ßa marche :
- **If** (si) : Si une condition est vraie, faites ceci.
- **Else** (sinon) : Sinon, faites cela.

Par exemple, v√©rifions si un nombre est pair ou impair.

````python
# v√©rification si un nombre est pair
nombre = 7 # Changez ce nombre pour tester
if nombre % 2 == 0:
    print(nombre, "est pair !")
else:
    print(nombre, "est impair !")
````


**Astuce** : `%` est l'op√©rateur modulo, qui donne le reste d'une division. Si `nombre % 2 == 0`, alors le nombre est pair.

### Testons avec plusieurs conditions

````python
# test avec plusieurs conditions
temperature = int(input("Devine la temp√©rature aujourd'hui (en ¬∞C) : "))

if temperature < 0:
    print("Brrr... Il g√®le ! ü•∂")
elif temperature < 10:
    print("Il fait froid. Mets un manteau ! üß•")
elif temperature < 20:
    print("Il fait frais. Un pull suffit. üß∂")
elif temperature < 30:
    print("Il fait bon, profite ! üå§Ô∏è")
elif temperature < 40:
    print("Ouh l√†, il fait chaud ! ü•µ")
else:
    print("C'est une canicule ! Reste √† l'ombre ! üî•")
````

### Boucles For
Les boucles **for** sont g√©niales pour r√©p√©ter quelque chose un certain nombre de fois. Par exemple, imprimons les nombres de 1 √† 5.

````python
for i in range(1, 5):  # De 1 √† 5
   print(i)
````

````python
for i in range(2, 5):
  print(i)
````

#### Boucle For - Codes amusants

- Evaluation d'un nombre

````python
# Evaluation d'un nombre
print("Entre 5 nombres. Je vais te dire ce que j'en pense !")

for i in range(5):
    nombre = int(input("Nombre #" + str(i + 1) + " : "))

    if nombre == 0:
        print("C'est z√©ro. Ni froid ni chaud. üòê")
    elif nombre < 0:
        print("Ouh l√†, c'est un nombre n√©gatif. ‚ùÑÔ∏è")
    elif nombre > 100:
        print("Wow, c‚Äôest un tr√®s grand nombre ! üöÄ")
    else:
        print("C'est un nombre positif. üëç")
````

- Contruction d'une pyramide

````python
# Construction d'une pyramide
print("Bienvenue dans le Constructeur de Pyramides Magiques !")

# Demander ton nom
nom = input("Bonjour ! Comment t'appelles-tu ? ")

# Demander le nombre de lignes
n = int(input(f"Super, {nom} ! Maintenant, entre un nombre entier pour le nombre de lignes de ta pyramide magique : "))

# Choisir un caract√®re pour la pyramide
caractere = input("Quel caract√®re veux-tu utiliser pour ta pyramide ? (par exemple, *, O, #) ")[0]

print("\nVoici ta pyramide :\n")

# Boucle for pour construire la pyramide
for i in range(1, n+1):
    print(" " * (n - i) + caractere * (2*i - 1))

# Commentaire amusant selon la taille
if n < 3:
    print("\nTa pyramide est petite mais charmante !")
elif n < 10:
    print("\nBelle pyramide ! Tu as bon go√ªt !")
else:
    print("\nIncroyable ! Tu es un architecte en herbe !")
````

NB: **Comment √ßa marche ?**

- Nom : Le programme te demande ton nom avec input() et l‚Äôutilise pour te parler.
- Nombre de lignes : Tu entres un nombre, converti en entier avec `int()`, pour d√©finir la hauteur de la pyramide.
- Caract√®re : Tu choisis un caract√®re, et le programme prend le premier que tu entres (avec [0]).

Boucle for : Elle r√©p√®te une action pour chaque ligne (`i` va de 1 √† n). Pour chaque ligne :
- `" " * (n - i)` ajoute des espaces pour centrer la pyramide.
- `caractere * (2*i - 1)` r√©p√®te ton caract√®re un nombre impair de fois (1, 3)
- Message final : Selon la taille de ta pyramide, tu re√ßois un commentaire amusant !

**R√©sultat visuel pour n = 3**:

Examinons ce qui est affich√© √† chaque it√©ration :

- Ligne 1 `(i = 1) : " " * (3 - 1) + "#" * (2*1 - 1)` ‚Üí " " + "#" = " #"
R√©sultat : # (2 espaces suivis d'un #).
- Ligne 2 `(i = 2) : " " * (3 - 2) + "#" * (2*2 - 1)` ‚Üí " " + "###" = " ###"
R√©sultat : ### (1 espace suivi de trois #).
- Ligne 3 `(i = 3) : " " * (3 - 3) + "#" * (2*3 - 1)`` ‚Üí "" + "#####" = "#####"
R√©sultat : ##### (0 espace suivi de cinq #).

### Boucles While (tant que)

Les boucles **while** r√©p√®tent tant qu'une condition est vraie. Soyez prudent, elles peuvent continuer ind√©finiment si vous ne faites pas attention !


````python
compteur = 1
while compteur <= 5:
    print(compteur)
    compteur = compteur + 1  # Augmente le compteur de 1
````


**Astuce** : N'oubliez pas d'augmenter le compteur dans une boucle while pour √©viter une boucle infinie !

#### Boucle While - Code Infini !

Pour l'interrompre :

Appuie sur Command + Ctrl et M en m√™me temps.
Puis rapidement, appuie deux fois sur la touche i (i i).
ou  Clique sur le bouton Arr√™ter (ic√¥ne de carr√© noir) √† c√¥t√© de la cellule en cours d'ex√©cution

Cela interrompt la cellule en cours d‚Äôex√©cution.

````python
while True:
    print("Coucou ! Je tourne en boucle pour toujours üåÄ")
````

NB:

- `while` signifie "tant que".

- `True` veut dire que la condition est toujours vraie.

Donc Python continue pour toujours √† ex√©cuter ce qu‚Äôil y a √† l‚Äôint√©rieur du bloc : ici, il affiche "Coucou !" encore et encore.!!!

#### Boucle While - Code Infini - Correction

````python
compteur = 0

while True:
    print("Coucou ! Je tourne en boucle üåÄ")
    compteur += 1

    if compteur == 10:
        print("Ouf ! Je m'arr√™te apr√®s 10 tours üéâ")
        break
````

NB: Une boucle peut √™tre infinie au d√©part (`while True:`).

On peut la contr√¥ler avec une condition et `break` pour qu‚Äôelle ne tourne pas vraiment √† l‚Äôinfini.

C‚Äôest important pour √©viter que le programme plante ou devienne incontr√¥lable.

#### Boucle While - Code amusant
Le d√© magique !

Le programme choisit un nombre secret entre 1 et 5.
Tu as 3 essais pour deviner.
√Ä chaque essai, le programme donne une r√©ponse dr√¥le, et si tu gagnes, tu es f√©licit√© comme un "sorcier" !

````python
# le d√© magique
import random  # module Python pour acc√®der √† des num√©ros al√©atoires

print("Bienvenue au D√© Magique ! üé≤")
print("Je pense √† un nombre entre 1 et 5. Devine-le !")

nombre_secret = random.randint(1, 5)
tentatives = 0

while tentatives < 3:
    guess = int(input("Ton guess (1 √† 5) : "))
    tentatives += 1

    if guess == nombre_secret:
        print("WOW ! T'es un sorcier ! ü™Ñ")
        break
    else:
        print("Hihi, rat√© ! üòú Essaie encore !")

if tentatives == 3 and guess != nombre_secret:
    print(f"Dommage ! Le nombre magique √©tait {nombre_secret} ! Rejoue ? üòÑ")
````

### Activit√© 2 : Cr√©er une fonction palindrome
√âcrivez une fonction qui v√©rifie si un mot est un palindrome (se lit de la m√™me mani√®re √† l'envers).

````python
# v√©rification pour savoir si un mot est un palindrome
def est_palindrome(mot):
    if mot == mot[::-1]:      # mot[::-1] inverse le mot en Python.
        print(mot, "est un palindrome !")
    else:
        print(mot, "n'est pas un palindrome.")

# Testez la fonction
est_palindrome("radar")
est_palindrome("bonjour")
est_palindrome("√©t√©")
````

### Jeu - Animal myst√®re

le code ci-dessous choisit un nombre al√©atoire, donne au joueur 3 chances de le deviner, et affiche des messages diff√©rents selon qu'il gagne ou perde (apr√®s 3 essais).

````python
import random

def jeu_animal_mystere():
    print("Bienvenue dans le jeu de l'animal myst√®re ! üêæ")

    while True: # Boucle principale pour rejouer
        print("\nJe pense √† un animal. Je vais te donner des indices, essaie de deviner !")

        # Indices pour l'animal myst√®re (par exemple, un chat)
        indices = [
            "Indice 1 : Cet animal a des moustaches.",
            "Indice 2 : Il aime courir apr√®s une balle ou une ficelle.",
            "Indice 3 : Il fait 'Miaou' !"
        ]

        animal_trouve = False # Variable pour savoir si l'animal a √©t√© trouv√©

        # Afficher les indices un par un
        for indice in indices:
            print(indice)
            reponse = input("Quelle est ta r√©ponse ? (Tape le nom de l'animal ou 'je ne sais pas') : ").lower()

            if reponse == "chat":
                print("üéâ Bravo ! Tu as trouv√©, c'est bien un chat ! üê±")
                animal_trouve = True
                break # Sort de la boucle des indices
            elif reponse == "je ne sais pas":
                print("Pas de souci, voici un autre indice !")
            else:
                print("Ce n'est pas √ßa, essaie encore avec le prochain indice !")

        # Si l'animal n'a pas √©t√© trouv√© apr√®s tous les indices
        if not animal_trouve:
            print("Oh non, tu n'as pas trouv√© ! L'animal myst√®re √©tait un chat ! üê±")

        # Demander si l'utilisateur veut rejouer
        rejouer = input("Veux-tu rejouer ? (oui/non) : ").lower()
        if rejouer != "oui":
            print("Merci d'avoir jou√© ! √Ä bient√¥t ! üëã")
            break # Sort de la boucle principale

# Lancer le jeu
jeu_animal_mystere()
````
