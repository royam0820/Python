# Setup - Import des biblioth√®ques pour le Machine Learning

```python
# üåüüåüüåü CLASSIFIEUR CHAT/CHIEN POUR ENFANTS üåüüåüüåü

from mlforkidsimages import MLforKidsImageProject
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import matplotlib
from google.colab import files
import matplotlib.font_manager as fm
```

```python
# üîë Cl√© secr√®te pour acc√©der √† au projet IA de classification de chiens et de chats
key = "0f5dfd60-98e7-11ef-ac02-79299d1233ac767cf40e-c40d-4a21-97ce-7062b15bf66f"
```

# Entra√Ænement du Mod√®le

```python
# üéì Entra√Ænement de l'IA
# ‚ö†Ô∏è Tu peux commenter ces lignes apr√®s la premi√®re ex√©cution pour gagner du temps !
print("\n" + "="*50)
print("üöÄ ENTRA√éNEMENT DE L'INTELLIGENCE ARTIFICIELLE...")
print("‚è≥ Patiente un peu pendant que je deviens super intelligent !")
print("="*50 + "\n")

myproject = MLforKidsImageProject(key)
myproject.train_model()  # C'est ici que la magie op√®re !

print("\n‚úÖ WOUAH ! Entra√Ænement termin√© ! Je suis pr√™t √† reconna√Ætre les animaux !")
```

# Choix d'une image

```python
# üì∏ T√âL√âCHARGEMENT DE L'IMAGE MYST√àRE
print("\n" + "="*50)
print("üì∏ T√âL√âCHARGE TON IMAGE MYST√àRE")
print("1. Clique sur 'Parcourir'")
print("2. S√©lectionne une photo de chat ou de chien")
print("3. Appuie sur 'Ouvrir'")
print("4. Attends que le nom du fichier apparaisse ci-dessous")
print("="*50 + "\n")

uploaded = files.upload()
IMAGE_NAME = list(uploaded.keys())[0]  # R√©cup√®re le nom du premier fichier upload√©

print(f"\nüéâ SUPER ! Image '{IMAGE_NAME}' t√©l√©charg√©e avec succ√®s !")
```

```python
# üîé Je vais analyser l'image comme un d√©tective !
print("\n" + "="*50)
print(f"üîç JE SCRUTE L'IMAGE : {IMAGE_NAME}")
print("üß† Je r√©fl√©chis tr√®s fort...")
print("="*50 + "\n")

```

# Pr√©diction

```python
prediction = myproject.prediction(IMAGE_NAME)
label = prediction["class_name"]
confidence = prediction["confidence"]

# üêõ CORRECTION D√âFINITIVE DU PROBL√àME DE POURCENTAGE
# V√©rification si la confiance est d√©j√† en pourcentage ou en d√©cimal
if confidence > 1.0:
    # Si la confiance est sup√©rieure √† 1, c'est qu'elle est d√©j√† en pourcentage
    display_confidence = confidence / 100.0  # Conversion en valeur d√©cimale
    percent_value = confidence  # Valeur en pourcentage pour l'affichage
else:
    # Si la confiance est inf√©rieure √† 1, c'est qu'elle est en d√©cimal
    display_confidence = confidence
    percent_value = confidence * 100  # Conversion en pourcentage

# üò∏üê∂ Pr√©paration des emojis rigolos pour le r√©sultat
if "cat" in label.lower():
    animal_text = "CHAT"
    animal_emoji = "üò∏"
    title_color = "#FF6B6B"  # Rose pour chat
elif "dog" in label.lower():
    animal_text = "CHIEN"
    animal_emoji = "üê∂"
    title_color = "#4ECDC4"  # Bleu pour chien
else:
    animal_text = "ANIMAL MYST√àRE"
    animal_emoji = "‚ùì"
    title_color = "#9B59B6"  # Violet pour inconnu

# üìä Cr√©ation d'un joli graphique avec le r√©sultat
plt.figure(figsize=(10, 8), facecolor='#FFFAF0')  # Fond cr√®me

# Affichage de l'image
img = np.array(Image.open(IMAGE_NAME))
plt.imshow(img)
plt.axis('off')

# üé® Titre stylis√© avec des couleurs
# SOLUTION : On utilise du texte simple pour √©viter les probl√®mes d'emojis dans les polices
title_text = f"{animal_text}\nPR√âDICTION: {label.upper()} \nCONFIANCE: {percent_value:.1f}%"

# Utilisation d'une police qui fonctionne bien dans Colab
font_name = 'DejaVu Sans'  # Cette police est toujours disponible dans Colab

plt.title(title_text,
          fontsize=20,
          fontweight='bold',
          color=title_color,
          pad=20)  # Espacement

# üñºÔ∏è Ajout d'un cadre d√©coratif
frame = plt.Rectangle((0,0), 1, 1, transform=plt.gca().transAxes,
                     fill=False, edgecolor="#FFD700", linewidth=8)
plt.gca().add_patch(frame)

# ‚ú® Ajout d'√©toiles magiques pour l'effet "wow" !
# On utilise des caract√®res simples pour √©viter les probl√®mes
plt.figtext(0.1, 0.9, "*", fontsize=40, color="#FFD700", alpha=0.7)
plt.figtext(0.9, 0.9, "*", fontsize=40, color="#FFD700", alpha=0.7)
plt.figtext(0.5, 0.1, "*", fontsize=30, color="#FFD700", alpha=0.7)

plt.tight_layout(pad=5)
plt.show()

# üì£ R√©sultat final dans la console (ici on peut utiliser les emojis sans probl√®me)
print("\n" + "="*50)
print("üéâ R√âSULTAT FINAL !")
print("="*50)
print(f"üñºÔ∏è Image analys√©e : {IMAGE_NAME}")
print(f"üîÆ Mon intelligence artificielle pense que c'est un...")
print(f"\n\tüåüüåüüåüüåü {animal_emoji} {label.upper()} {animal_emoji} üåüüåüüåüüåü\n")
print(f"üíØ Niveau de confiance : {percent_value:.1f}%")

# Message d'√©valuation de la confiance
if display_confidence > 0.8:
    print("üòé IMPRESSIONNANT !")
elif display_confidence > 0.6:
    print("üëç PAS MAL !")
else:
    print("ü§î HUMM... J'H√âSITE UN PEU")

print("="*50 + "\n")

# üíå Message sp√©cial selon l'animal
if "cat" in label.lower():
    print(f"{animal_emoji} Miaou ! Les chats sont les ma√Ætres du monde !")
    print("üêæ Conseils pour les chats :")
    print("   - Offre-lui des croquettes d√©licieuses")
    print("   - Pr√©pare un carton pour qu'il puisse se cacher")
    print("   - Chatouille-lui le ventre (mais attention aux griffes !")
elif "dog" in label.lower():
    print(f"{animal_emoji} Ouaf ! Les chiens sont les meilleurs amis des humains !")
    print("ü¶¥ Conseils pour les chiens :")
    print("   - Lance-lui une balle pour jouer")
    print("   - Donne-lui des caresses sur le ventre")
    print("   - Emm√®ne-le en promenade chaque jour")
else:
    print(f"{animal_emoji} Wow ! Tu as trouv√© un animal rare !")
    print("üîç Essaye avec une autre photo pour voir si je le reconnais mieux")
    print("üåç Peut-√™tre as-tu d√©couvert une nouvelle esp√®ce ?")

print("\nüí° Astuce : Tu peux relancer ce programme pour tester d'autres images !")
print("üéÆ Amuse-toi bien avec ton IA !")
```
