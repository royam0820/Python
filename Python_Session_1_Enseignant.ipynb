### Variables et types de donn√©es
Les variables sont comme des bo√Ætes qui contiennent des informations. Elles peuvent stocker :

- **Boul√©ens** (Boolean): Vrai ou Faux (True ou False)


````python
# Exemples
nom = "Alex"  # Cha√Æne de caract√®res - string
age = 12      # Entier - integer
taille = 1.42  # Flottant - float
est_jeune = True  # boul√©en

print(nom)
print(age)
print(taille)
print(est_jeune)
````

````python
# prompt: types de donn√©es dans Python, donne moi les principaux types

# Types de donn√©es principaux en Python

# 1. Nombres entiers (int)
entier = 10
print(entier, type(entier))  # Affiche <class 'int'>

# 2. Nombres √† virgule flottante (float)
flottant = 3.14
print(flottant, type(flottant))  # Affiche <class 'float'>

# 3. Cha√Ænes de caract√®res (str)
chaine = "Bonjour le monde !"
print(chaine, type(chaine))  # Affiche <class 'str'>

# 4. Bool√©ens (bool)
booleen_vrai = True
booleen_faux = False
print(booleen_vrai, type(booleen_vrai))  # Affiche <class 'bool'>

# 5. Listes (list) - collections ordonn√©es et modifiables
liste = [1, 2, 3, "a", "b"]
print(liste, type(liste))  # Affiche <class 'list'>

# 6. Tuples (tuple) - collections ordonn√©es et immuables
tuple_ex = (1, 2, 3, "a", "b")
print(tuple_ex, type(tuple_ex)) # Affiche <class 'tuple'>

# 7. Dictionnaires (dict) - collections de paires cl√©-valeur
dictionnaire = {"nom": "Alice", "age": 30}
print(dictionnaire, type(dictionnaire)) # Affiche <class 'dict'>

# 8. Ensembles (set) - collections non ordonn√©es d'√©l√©ments uniques
ensemble = {1, 2, 2, 3}  # L'√©l√©ment 2 est pr√©sent une seule fois
print(ensemble, type(ensemble)) # Affiche <class 'set'>
````

### Math√©matiques de base
Python peut faire des calculs comme une calculatrice ! Utilisez ces symboles :
- `+` pour l'addition
- `-` pour la soustraction
- `*` pour la multiplication
- `/` pour la division
- `**`pour l'exposant
- `//`division enti√®re

````python
# Essayez quelques calculs
print(5 + 3)   # Addition
print(10 * 2)  # Multiplication
print(10 / 3)  # Division
print(10 // 3) # Division enti√®re
print(10 % 3)  # Modulo, ce qui reste apr√®s avoir divis√© 10 par 3
````

#### Formatage des chiffres d√©cimaux

````python
# formater un r√©sultat d√©cimal √† deux chiffres apr√®s la virgule - deux exemples
print(f"{10/3:.2f}")  # ex 1
print(round(10/3, 2)) # ex 2
````

> ü§ñ **Demande √† l'IA** :

> ‚û°Ô∏è donne moi les r√®gles de calcul dans Python

````python
# prompt: donne moi les r√®gles de calcul dans Python

# Exemple 1 : Parenth√®ses prioritaires
resultat = (10 + 5) * 2  # (10 + 5) est √©valu√© en premier, puis multipli√© par 2
print(resultat)  # Affiche 30

# Exemple 2 : Exposant avant multiplication
resultat = 2 ** 3 * 2  # 2 ** 3 (2 √† la puissance 3) est √©valu√© en premier, puis multipli√© par 2
print(resultat)  # Affiche 16

# Exemple 3 : Multiplication avant addition
resultat = 5 + 3 * 2  # 3 * 2 est √©valu√© en premier, puis ajout√© √† 5
print(resultat)  # Affiche 11

# Exemple 4 : Division et multiplication de gauche √† droite
resultat = 10 / 2 * 5  # 10 / 2 est √©valu√© en premier (5), puis multipli√© par 5
print(resultat)  # Affiche 25

# Exemple 5 : Division enti√®re
resultat = 10 // 3
print(resultat) # Affiche 3

# Exemple 6: Modulo (reste de la division)
resultat = 10 % 3
print(resultat) # Affiche 1


# Utilisation de parenth√®ses pour contr√¥ler l'ordre
resultat = 10 / (2 * 5) # Ici, (2 * 5) est √©valu√© en premier
print(resultat)  # Affiche 1.0
````

Version de l'histoire modifi√©e pour prendre en compte les articles

````python
def genre_article(nom):
  """D√©termine le genre d'un nom (masculin ou f√©minin) et retourne l'article appropri√©."""
  # Liste de mots g√©n√©ralement f√©minins (non exhaustive)
  feminins = ["maison", "voiture", "table", "chaise", "fille", "femme", "chatte"]
  if nom.lower() in feminins:
    return "la"
  else:
    return "le"

# Demande les informations √† l'utilisateur
nom = input("Quel est ton nom ? ")
animal = input("Quel animal : ")
nourriture = input("Quel est ton plat pr√©f√©r√© ? ")
lieu = input("Quel lieu ? ")

# Construction de l'histoire
histoire = f"Il √©tait une fois, {nom} qui a vu {genre_article(animal)} {animal} manger une {nourriture} dans un {lieu} !\n"
histoire += f"{genre_article(animal).capitalize()} {animal} a dit : 'Tu veux te joindre √† moi ?'"

# Affichage de l'histoire
print(histoire)
````
