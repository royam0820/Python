## POO - Les Concepts


**Les concepts**

- **Classe** : Un mod√®le ou un plan pour cr√©er des objets (ex. : un plan pour construire un jouet).
- **Objet** : Une instance d‚Äôune classe, comme un jouet fabriqu√© √† partir du plan.
- **H√©ritage** : Une classe peut "h√©riter" des caract√©ristiques d‚Äôune autre classe, comme un chiot qui h√©rite des traits d‚Äôun chien.

### Voyant cela en code !

`````python
# La classe m√®re
class Dog:
    """
    Represents a dog with a name and breed.
    """
    def __init__(self, name, breed):
        """
        Initialise un nouvel objet Dog.

        Args:
            name (str): Le nom du chien.
            breed (str): La race du chien.
        """
        self.name = name
        self.breed = breed

    def bark(self):
        """
        Simule les aboiements du chien.

        Returns:
            str: Une cha√Æne repr√©sentant l'aboiement du chien.
        """
        return f"{self.name} barks: Woof woof!"

# Example d'utilisation (instanciation):
my_dog = Dog("Buddy", "Golden Retriever")   # nom et race du chien
print(my_dog.bark())                        # ex√©cution de la m√©thode bark
`````

`````python
# La classe fille
class GoldenRetriever(Dog):
    """
    Repr√©sente un Golden Retriever, h√©ritant de la classe Dog.
    """
    def __init__(self, name):
        """
        Constructeur pour la classe GoldenRetriever.

        Args:
            name: le nom du Golden Retriever.
        """
        super().__init__(name, "Golden Retriever")

    def fetch(self):
        """
        Une m√©thode sp√©cifique aux Golden Retrievers pour rapporter la balle.
        Returns:
            Une cha√Æne repr√©sentant l'action de r√©cup√©ration.
        """
        return f"{self.name} is fetching the ball!"

# Example usage of inheritance:
my_golden = GoldenRetriever("Goldie")
print(my_golden.bark())  # Inherited method
print(my_golden.fetch()) # Specific method
`````
